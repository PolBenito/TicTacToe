```python
import csv
import numpy as np
import pandas as pd

# bbdd = pd.read_csv('bbdd_primitiva.csv', delimiter = '\t')

# Pasar la BBDD a una lista
import csv

v_bbdd = []
count = 0

with open('bbdd_primitiva.csv') as f:
    reader = csv.reader(f)
    for row in reader:        
        if count != 0:
            linea = row[1:]
            
            tablero = [
                [linea[0], linea[1], linea[2]],
                [linea[3], linea[4], linea[5]],
                [linea[6], linea[7], linea[8]]
            ]
            
            v_bbdd.append(tablero)

        count += 1
        
# Crear nuevas formas de la matriz
for x in v_bbdd:
    bbdd = np.fliplr(x)
    v_bbdd.append(bbdd)

# from sklearn.utils import shuffle
# v_bbdd = pd.DataFrame(shuffle(v_bbdd))
# v_bbdd = pd.DataFrame(v_bbdd)

# bbdd.head()
v_bbdd
```

```python
from pandas import *
import numpy as np

v_tab = np.array([['a', 'b', 'c'],
                  ['d', 'e', 'f'],
                  ['g', 'h', 'i']])

print (DataFrame(v_tab))
print ()
print ('Tabla al revés')
print (np.fliplr(v_tab))
print ()
print ('Tabla bocabajo')
print (np.flipud(v_tab))
print ()
print ('Tabla bocabajo y al revés')
print (np.fliplr(np.flipud(v_tab)))
print ()
print ('Tabla rotación 90º sentido antihorario')
print (np.rot90(v_tab))
print ()
print ('Tabla rotación 90º sentido horario')
print (np.fliplr(np.flipud(np.rot90(v_tab))))
print ()
print ('Tabla ordenada por columnas')
print (np.flipud(np.rot90(v_tab)))
print ()
print ('Tabla rotación 90º sentido antihorario al revés')
print (np.fliplr(np.rot90(v_tab)))
```
