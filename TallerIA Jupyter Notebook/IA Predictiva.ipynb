```python
import csv
import pandas as pd
from sklearn.utils import shuffle


bbdd = pd.read_csv('bbdd_amb_canvis_h_v.csv')

bbdd = pd.DataFrame(shuffle(bbdd))
bbdd.head()
```

```python
comprova_aixo = input('Quina es la jugada que vols comprovar: ')
nova_llista = []
for i in comprova_aixo:
    nova_llista.append(int(i))
comprova_aixo = nova_llista
print (list(comprova_aixo))
```

```python
#Importem les llibreries Sklearn per tal de calcular l'acurracy i ttictoc.

from sklearn.metrics import accuracy_score
from ttictoc import TicToc                          #Serveix per calcular el temps de codi entre dos marcadors.

t = TicToc()                                        #TicToc("name").

#Funció per determinal els persentatges d'entrenament y test:
#------------------------------------------------------------------------------------------------------------------------------#
def percentage_test(percentage, data):              #"percentatge": dades a entrenar en tant per cent. "data": base de dades.
    llindar = int((data.shape[0]/100)*percentage)
    return llindar                                  #La funció retorna el nombre de la última fila abans de la part de test.
#------------------------------------------------------------------------------------------------------------------------------#


#Funció per aplicar els métodes de classificació:
#------------------------------------------------------------------------------------------------------------------------------#
def classificacio(nom, dades, clf, llindar):        #"nom": nom del algoritme. "dades": base de dades. "clf": Classificador.
                                                    #"llindar": límit de les dades d'entrenament.
        
    t.tic()                                         #Primer marcador de temps Tic().
    print(nom)
    print()
    
    #Seleccionem primer els atributs (les dades que fem servir per predir) i les classes (el que volem predir).
    #el .fit tansols prepara les dades per introduir-les al model de classificació
    clf.fit(dades.iloc[:llindar,1:], dades.iloc[:llindar,0])
    
 
    #Definim les dades test que predim ("preds") i les dades test que son ("test").
    preds = pd.Series(clf.predict(dades.iloc[llindar:,1:]), name='preds')#li estàs dient 
    #"prediu-me què hi ha a la columna -1"
    test = pd.Series(dades.iloc[llindar:,0],name='reales')     #li estàs dient "diga'm què hi ha realment a la columna -1"
    test.index = range((dades.shape[0]) - llindar)
    
    #ara fem un test random amb una dada que decidim nosaltres pk puedo, que apareix indexat amb el 0
    global comprova_aixo
    test_random = pd.Series(clf.predict([comprova_aixo]))
    
    acc = accuracy_score(test, preds)               #Es calcula l'"accuracy" i la matriu de covariança.
    acc = acc*100
    
    cov = pd.crosstab(test, preds)
    t.toc()                                         #Segon marcador de temps Toc().
    
    #Mostrem les dades.
    #print("test random", test_random)
    
    print('acc:', round(acc, 2), "%")               #Accuracy del conjunt.
    print('temps: ', round(t.elapsed, 3), 's')      #Temps d'execució del entrenament.
    print()
    print(cov)                                      #Matriu de covariança.
    return(test_random)
#-------------------------------------------------------------------------------------------------------------------
```

```python
from sklearn.neighbors import KNeighborsClassifier

#Utilitzem les funcions creades per a entrenar la base de dades:

llindar = percentage_test(90, bbdd)                      #90 % de la base de dades

k = 1                                                       #Paràmetre k del classificador KNN
valor = classificacio('kNN (k=1)', bbdd, KNeighborsClassifier(k), llindar)
print("test random", valor)
```

```python
#Importem l'algoritme Random Forest de la llibreria sklearn.
from sklearn.ensemble import RandomForestClassifier

#Utilitzem les funcions creades per a entrenar la base de dades:

llindar = percentage_test(90, bbdd)          #70 % de la base de dades.
n = 100                                         #Paràmetre (nombre d'estimadors) de Random Forest.
valor = classificacio('Random Forest', bbdd, RandomForestClassifier(n_estimators=n), llindar)
```
